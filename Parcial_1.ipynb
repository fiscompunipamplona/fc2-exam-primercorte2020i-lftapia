Como podemos apreciar la kurtosis para todos los datos son mas lejanos tanto para la derecha como para la izquierda de la distribución de datos esto significa que con respecto a los valores esperedos son un poco aceptables

```
#Covarianza y Correlacion
x = data[:,0] #Periodo orbital
y = data[:,3] #Angulo de inclinacion
xy = x*y
p = sum(x)/len(x)
t = sum(y)/len(y)
pp = sum ( (x-p)*(x-p) )/(len(x)-1)
tt = sum ( (y-t)*(y-t) )/(len(y)-1)
covarianza=(sum(xy)/len(xy))-((sum(x)/len(x)) * (sum(y)/len(y)))
correlaccion=covarianza/(sqrt(pp)*sqrt(tt))
print("La covarianza entre el periodo orbital y la excentrincidad es de:",format(covarianza,'.2f'))
print("\nLa correlación entre el periodo orbital y la excentrincidad es de:",format(correlaccion,'.2f'))
```

Como se aprecia estan variables las cuales son el periodo orbital y el angulo de inclinacion de los exoplanetas la covarianza es positiva lo que significa que la relacion es directamente proprcional es decir a mayor periodo orbital los exoplanetas se estan inclinando mas.
Para la correlación entre estas dos son muy dependientes comprobando lo dicho anteriormente.
